```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = "/content/Unemployment in India.csv"
df = pd.read_csv(file_path)
```

```python
# Display basic information about the dataset
print(df.info())
```

```python
# Display the first few rows of the dataset
print(df.head())
```

```python
# Check for missing values
print("Missing Values:")
print(df.isnull().sum())
```

```python
# Replace ' Date' with the correct column name representing time-related information
time_column_name = ' Date'  # Note the leading space in the column name
df[time_column_name] = pd.to_datetime(df[time_column_name])

# Set the time column as the index
df.set_index(time_column_name, inplace=True)
```

```python
# Explore the distribution of unemployment rates
plt.figure(figsize=(12, 6))
sns.histplot(df[' Estimated Unemployment Rate (%)'], bins=20, kde=True)
plt.title('Distribution of Unemployment Rates in India')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Plot the unemployment rate over time
plt.figure(figsize=(20, 6))
plt.plot(df[' Estimated Unemployment Rate (%)'], marker='o', linestyle='-', color='b')
plt.title('Unemployment Rate in India Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.show()
```

```python
# Correlation
df.corr()
```

```python
# Correlation with other variables with matrix
correlation_matrix = df.corr()
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
